```python
from geetools.tb import *
import matplotlib.pyplot as plt
```

```python
fecha1 = parse("2022-04-05") 
fecha2 = fecha1 + pd.Timedelta("5D")

tmp= TB(config_file='config_gbv.ini',device_name='Termopares cafe')
sensores = ["Thermocouple_1","Thermocouple_10","Thermocouple_11",
            "Thermocouple_12","Thermocouple_2","Thermocouple_3",
            "Thermocouple_4","Thermocouple_5","Thermocouple_6",
            "Thermocouple_7","Thermocouple_8","Thermocouple_9"]
cafe = pd.concat([tmp.get_df(key=sensor,start_datetime=fecha1,end_datetime=fecha2).resample('1min').mean() for sensor in sensores],axis=1)
```

```python
cafe.plot()
```

```python

fecha11 = parse("2022-04-07 15:30") 
fecha22 = fecha11 + pd.Timedelta("4H")


fig, ax = plt.subplots(figsize=(18,6))


ax.set_xlim(fecha11,fecha22)
for sensor in sensores:
    ax.scatter(cafe[sensor].index,cafe[sensor])
ax.set_ylim(30,38)

ax.grid()
```
